```python
from openai import OpenAI
import os

# Point to the local server
client = OpenAI(base_url="http://localhost:1234/v1", api_key="lm-studio")


def make_questions(data):
    completion = client.chat.completions.create(
        model="lmstudio-community/Meta-Llama-3-8B-Instruct-GGUF",
        messages=[
            {
                "role": "system",
                "content": """You are a university of Northampton AI Course bot, made to anwer student queries around the course and uni! There is some loose data around the course, can you please generate some useful questions using the data provided generate ony 20 questions""",
            },
            {
                "role": "user",
                "content": data
            },
        ],
        temperature=0.1,
    )
    return completion.choices[0].message.content
```

```python
import os
import docx2txt
import math
```

```python
docs = os.listdir('.\AI_COURSE_DATA')
```

```python
CONTEXT_WINDOW = 200
```

```python
intents_dict = {}
for i in docs:
    content = docx2txt.process(os.path.join('.\AI_COURSE_DATA',i))
    stripped_content = content.strip()
    divided_content = stripped_content.split(' ')
    no_of_words = len(divided_content)
    batches = math.ceil(no_of_words/CONTEXT_WINDOW)
    last_token = CONTEXT_WINDOW
    start_token = 0
    print('\n')
    print(i)
    print("=============================================================================================")
    print("number_of_words", {no_of_words})
    print("number_of_batches", {batches})
    intents_dict[i] = []
    for batch_no in range(1,batches+1):
        if batch_no == batches:
            print('----------------------------------------------------------------------------------------------')
            print(start_token)
            print(batch_no)
            #print(' '.join(divided_content[start_token : ]))
            batch_joined = ' '.join(divided_content[start_token : ])
            generated_questions = make_questions(batch_joined)
            print(generated_questions)
            intents_dict[i].append({'input' : batch_joined, 'output' : generated_questions})
           
        else:
            print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
            print(start_token)
            print(last_token)
            #print(' '.join(divided_content[start_token:last_token]))
            batch_joined = ' '.join(divided_content[start_token:last_token])
            generated_questions = make_questions(batch_joined)
            print(generated_questions)
            start_token += CONTEXT_WINDOW
            last_token += CONTEXT_WINDOW
            intents_dict[i].append({'input' : batch_joined, 'output' : generated_questions})
    print("=============================================================================================")
```

```python
def make_answers(data):
    completion = client.chat.completions.create(
        model="lmstudio-community/Meta-Llama-3-8B-Instruct-GGUF",
        messages=[
            {
                "role": "system",
                "content": """You are a University of Northampton AI Course bot, designed to answer student queries regarding the course and university. You have a dictionary pair of data and some information around it. Your task is to summarize a single answer for all of the questions provided.""",
            },
            {
                "role": "user",
                "content": data
            },
        ],
        temperature=0.1,
    )
    return completion.choices[0].message.content
```

```python
dict_que_ans_intent = {}
for i in intents_dict.keys():
    dict_que_ans_intent[i] = []
    for j in intents_dict[i]:
        answer = make_answers(str(j))
        print(answer)
        dict_que_ans_intent[i].append({'question' : j , 'answer' : answer})        
```

```python
with open('lala.txt','w') as t:
    for i in intents_dict.keys():
        t.writelines(f'intent: {i}')
        for j in dict_que_ans_intent[i]:
            print(j)
            t.write(f"questions: {j['question']}")
            t.writelines(f"Answers: {j['answer']}")
```
